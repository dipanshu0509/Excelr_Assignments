## 1.Data Cleaning and Preparation

```python
# Check for missing values
data.isnull().sum()
```

```python
# Handle missing values (e.g., fill with mean for simplicity)
num_cols=['LB', 'AC', 'DS', 'DP','MLTV', 'Width', 'Tendency', 'NSP']
for col in num_cols:
    data[col].fillna(data[col].mean(), inplace=True)
# Verify if missing values are handled
print(data.isnull().sum())
```

### Converting NSP and Tendency to Categorical

```python
data['NSP']=data['NSP'].astype('category')
data['Tendency']=data['Tendency'].astype('category')
data.dtypes
```

```python
#printing the boxplot

plt.figure(figsize=(14,10))
sns.boxplot(data)
plt.title("Numerical Variable Box plot")
plt.show()
```

**Treating the outliers**

```python
#defining a function to treat the outliers

def Outliers_data(data,columns):
    for col in columns:
        q1=data[col].quantile(0.25)
        q3=data[col].quantile(0.75)
        IQR=q3-q1
        lb=q1-1.5*IQR
        ub=q3+1.5*IQR
        data=data[(data[col]>=lb)&(data[col]<=ub)]
    return data   
columns_to_treat=['LB','AC','FM','UC','DL','DP','ASTV','MSTV','ALTV','MLTV','Width']
data_cleaned=Outliers_data(data, columns_to_treat)

print("cleaned shaped Dataframe:", data_cleaned.shape)
```

## Conclusion and Summary

### Key Insights and Patterns from the Cardiotocographic Dataset

1. **Fetal Heart Rate Trends**:
   - The Baseline Fetal Heart Rate (LB) typically ranges between 120-140 bpm, with variations indicating potential fetal distress. Instances of lower LB values could suggest complications requiring further monitoring.

2. **Correlation Between Measurements**:
   - Positive correlations exist between Accelerations (AC) and Fetal Movements (FM), indicating that increased fetal activity often aligns with healthy heart rates. Conversely, a negative correlation between LB and abnormal conditions (DP) suggests that higher percentages of abnormal readings correlate with lower baseline heart rates.

3. **Normal vs. Abnormal Status**:
   - The dataset shows a predominance of normal status (NSP = 1), but the presence of abnormal readings highlights the need for close examination of those cases, particularly where LB is low or AC and FM are infrequent.

### Impact on Decision-Making

- **Clinical Monitoring**: Continuous monitoring of LB, AC, and FM is essential for assessing fetal health. Anomalies in these parameters should prompt further evaluation.
- **Risk Stratification**: Identifying trends and correlations can help healthcare providers detect at-risk pregnancies early, allowing for timely interventions.
- **Predictive Modeling**: Future analyses could leverage machine learning to predict fetal distress based on historical data patterns, enhancing clinical decision-making and patient care strategies.
