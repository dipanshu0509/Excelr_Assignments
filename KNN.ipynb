```python
def remove_outliers(data):
  numerical_columns = data.select_dtypes(include = {'int64','float64'})
  for i in numerical_columns:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    UL = Q3 + (1.5 * IQR)
    LL = Q1 - (1.5 * IQR)
    data = data[(data[i] >= LL) & (data[i] <= UL)]
  return data

out_data = remove_outliers(zoo)
out_data
```

```python
import numpy as np
from sklearn.model_selection import KFold ,cross_val_score,GridSearchCV
```

```python
n_neighbors = np.array(range(0,80))
parameters = dict(n_neighbors = n_neighbors)
parameters
```

 interview questions:

 1. What are the key hyperparameters in KNN?

The key hyperparameters in KNN are:

- Number of Neighbors (k): The number of nearest neighbors that the algorithm considers when classifying a point. Choosing the right ( k ) is critical, as a smaller ( k ) makes the model more sensitive to noise, while a larger ( k ) can smooth out distinctions between classes.

- Distance Metric: Determines how the distance between points is calculated. Common metrics include Euclidean, Manhattan, and Minkowski distances. The choice of distance metric affects how the algorithm identifies "neighbors."

- Weight Function: Specifies whether all points are weighted equally or closer points should have more influence. Options include:
  - Uniform weighting: All neighbors are weighted equally.
  - Distance weighting: Closer neighbors have more influence.

- Algorithm for Neighbor Search: Determines how neighbors are searched, affecting performance:
  - Brute-force: Computes distance to all points; suitable for small datasets.
  - KD-tree or Ball-tree: Efficient for larger datasets, especially in lower-dimensional data.

- Leaf Size (for KD-tree/Ball-tree): Controls the number of points in each leaf node, impacting the speed and memory efficiency of the tree.

 2. What distance metrics can be used in KNN?

Several distance metrics can be used in KNN, and the choice depends on the data characteristics:

- Euclidean Distance: Measures the straight-line distance between two points. It is the most common choice for continuous, high-dimensional data.

- Manhattan Distance: Measures the distance along the axes of the feature space. It is often useful when features have different scales or when the data is more "grid-like" (e.g., text data).
    
- Minkowski Distance: A generalization of both Euclidean and Manhattan distances, controlled by a parameter ( p ):
  - ( p = 1 ) gives Manhattan distance.
  - ( p = 2 ) gives Euclidean distance.
  - Higher ( p ) values place more emphasis on the dimensions with larger differences.

- Hamming Distance: Useful for categorical data, it counts the number of feature differences between two points. 

- Cosine Similarity:Measures the cosine of the angle between two vectors, making it useful for text data or cases where orientation is more important than magnitude.

These distance metrics allow KNN to be applied across a range of datasets with different structures.
