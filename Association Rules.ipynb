```python
import warnings
import pandas as pd
warnings.filterwarnings('ignore')
data=pd.read_excel("Online retail.xlsx")
```

```python
data
```

**DATA PREPROCESSING**

```python
data=data[data.columns[0]].str.split(',',expand=True)
data.head()
```

```python
data[data.duplicated()]
```

```python
data=data.drop_duplicates()
```

```python
data.isna()
```

```python
data.dropna(axis=1, how='all', inplace=True)
data.dropna(axis=0, how='all', inplace=True)
```

```python
data.head()
```

```python
cleaned_data = data.fillna('')
cleaned_data.head()
```

```python
transactions = cleaned_data.apply(lambda row: [item for item in row if item != ''], axis=1).tolist()
transactions[0:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
##Fitting and transform the transaction into an array format
te_ary = te.fit(transactions).transform(transactions)
```

```python
# Convert the array into a DataFrame
basket = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
basket.head()
```

```python
data.to_csv("data.csv",index=False)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Apply the Apriori algorithm
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

**Analysis and Interpretation**

```python
def format_itemset(itemset):
    return ', '.join(sorted(list(itemset)))
```

```python
rules['antecedents_formatted'] =  rules['antecedents'].apply(format_itemset)
rules['consequents_formatted'] =  rules['consequents'].apply(format_itemset)
rules
```

```python
print("Top 10 association rules")
rules[['antecedents_formatted', 'consequents_formatted','support','confidence','lift']]
```

```python
antecedents = rules['antecedents_formatted'].value_counts().head(5)
consequents = rules['consequents_formatted'].value_counts().head(5)
```

```python
print("Top 5 Antecedents:")
antecedents
```

```python
print("Top 5 Consequents:")
consequents
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(rules['lift'], bins=20, kde=True)
plt.title('Distribution of Lift Values')
plt.xlabel('Lift')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(rules['support'], rules['confidence'], alpha=0.5, color='blue')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
high_lift_rules = rules[(rules['lift'] > 1.5) & (rules['support'] > 0.01)]
high_lift_rules[['antecedents_formatted', 'consequents_formatted', 'support', 'confidence', 'lift']]
```

```python
rules['antecedent_size'] = rules['antecedents'].apply(len)
rules['consequent_size'] = rules['consequents'].apply(len)
```

```python
avg_antecedent_size = rules['antecedent_size'].mean()
avg_consequent_size = rules['consequent_size'].mean()
```

```python
print(f"Average antecedent size: {avg_antecedent_size:.2f}")
```

```python
print(f"Average consequent size: {avg_consequent_size:.2f}")
```

```python
all_products = rules['antecedents'].apply(list) + rules['consequents'].apply(list)
product_counts = pd.Series([item for sublist in all_products for item in sublist]).value_counts()
product_counts
```

**Observations**


1. High Lift Values: The rules with high lift values indicate strong associations between the products. For example, the combination of "frozen vegetables" and "ground beef" with "spaghetti" has a high lift, suggesting that these items are often purchased together.

    
2. Common Products:"Mineral water" frequently appears as a consequent, indicating it is a common follow-up purchase with other items.


3. Rule Sizes:The average antecedent size is 2, and the average consequent size is 1, indicating that most rules involve two products leading to one consequent product.

Interview Questions:

1.What is lift and why is it important in Association rules?

A)Lift, a key concept in association rule mining, measures the surprisingness or strength of an association rule.
  It indicates how much greater the probability of the right-hand side (consequent) is in the presence of the left-hand side       (antecedent) compared to its probability in the absence of the antecedent.
  A higher lift value represents a more significant association, making it more important in decision-making processes.
  Lift helps identify the most relevant relationships between items, allowing businesses to optimize marketing strategies,         product recommendations, and customer segmentation.

2.What is support and Confidence. How do you calculate them?

A)Support and confidence are fundamental concepts in decision-making and machine learning, particularly in classification and     regression tasks.
 
 Support refers to the proportion of instances in the training dataset that belong to a certain class or category.
  
  Confidence, on the other hand, represents the certainty or reliability of a model's prediction.

   Calculation:
    Support : counting the number of instances in each class and dividing by the total number of instances.

    Confidence : calculated by dividing the number of transactions containing both itemsets by the number of transactions           containing the first itemset.
